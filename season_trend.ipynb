```python

    




lines = []
class category:
    def __init__(self, name, count):
        self.name = name
        self.count = count
with open('2017Trans.csv', 'rb') as f:
     lines = [l.decode('utf8', 'ignore') for l in f.readlines()]
category_dic = {}
for line in lines[1:]:
    arr = line.split(',')
    if arr[8] == 'TRUE' and arr[4] and arr[5] and arr[3]:
        if arr[3] not in category_dic:
            category_dic[arr[3]] = 1
        else:
            category_dic[arr[3]] += 1
l = []
for key in category_dic:
    l.append(category(key, category_dic[key]))
l = sorted(l, key = lambda x : x.count, reverse = True)
for i in l:
    print(i.name +': ' + str(i.count))

```

```python
lines = []
with open('trans_final.csv', 'rb') as f:
     lines = [l.decode('utf8', 'ignore') for l in f.readlines()]
category_pool = {}
count = 0
for line in lines[1:]:
    arr = line.split(',')
    strs = arr[0].split('-')
    if len(strs) < 2:
        continue
    year = strs[0]
    month = strs[1]
    if arr[7] not in category_pool:
        category_pool[arr[7]] = [0 for i in range(4)]
    if year == '2017':
        month = int(month)
        if month < 4:
            category_pool[arr[7]][0] += 1
        elif month < 7:
            category_pool[arr[7]][1] += 1
        elif month < 10:
            category_pool[arr[7]][2] += 1
        else:
            category_pool[arr[7]][3] += 1
```

```python
import matplotlib.pyplot as plt
size = [0,1,2,3]
fig, ax = plt.subplots(figsize=(9, 7), linewidth=20)
ax.set(xlabel='Seasons',ylabel='Numver of occurences',title='Count of event occurences in each season in 2017')
for key in category_pool:
    ax.plot(size, category_pool[key],'-o',label=key)
ax.legend()
ax.grid()
plt.show()
```
